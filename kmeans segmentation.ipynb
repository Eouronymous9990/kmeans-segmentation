```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img_ = cv2.imread(r"C:\Users\zbook 17 g3\Desktop\CV PROJECTS\stsci-nh-p2101a-f-1920x1080.png")
img_shape = img_.shape
img_flat = img_.reshape((-1, 3))
img_float32 = np.float32(img_flat)
k = 2
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.90)
retval, bestLabels, centers = cv2.kmeans(img_float32, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
centers = np.uint8(centers)
reult_image = centers[bestLabels.flatten()]
reult_image = reult_image.reshape(img_shape)

cv2.imwrite("processed_image.png", reult_image)

processed_img = cv2.imread("processed_image.png")
processed_img_rgb = cv2.cvtColor(processed_img, cv2.COLOR_BGR2RGB)
```

```python

plt.imshow(processed_img_rgb)
plt.axis("off")
plt.show()
```
